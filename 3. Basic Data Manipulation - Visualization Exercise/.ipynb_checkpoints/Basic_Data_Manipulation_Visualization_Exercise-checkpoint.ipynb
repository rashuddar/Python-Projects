### 33.  For each occupation present the percentage of women and men

```python
users.head()
```

```python
grouped = pd.pivot_table(data= users,
                        index= 'occupation',
                        columns = 'gender',
                        aggfunc='size')

grouped['total'] = grouped.sum(axis=1)
grouped['percentage_female']= (grouped['F']/grouped['total'])*100
grouped['percentage_male']= (grouped['M']/grouped['total'])*100

grouped
```

### 66. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
data
```

```python
data.index = pd.to_datetime(data.index)

start_date = pd.Timestamp('1961-01-02')

end_date = pd.Timestamp(start_date) + pd.DateOffset(weeks=52)
weekly_data = data[start_date:end_date]

weekly_stats = weekly_data.resample('W').agg({
    'RPT': ['min', 'max', 'mean', 'std'],
    'VAL': ['min', 'max', 'mean', 'std'],
    'ROS': ['min', 'max', 'mean', 'std'],
    'KIL': ['min', 'max', 'mean', 'std'],
    'SHA': ['min', 'max', 'mean', 'std'],
    'BIR': ['min', 'max', 'mean', 'std'],
    'DUB': ['min', 'max', 'mean', 'std'],
    'CLA': ['min', 'max', 'mean', 'std'],
    'MUL': ['min', 'max', 'mean', 'std'],
    'CLO': ['min', 'max', 'mean', 'std'],
    'BEL': ['min', 'max', 'mean', 'std'],
    'MAL': ['min', 'max', 'mean', 'std']
})

print(weekly_stats)
```

### 88. Calculate total sales data (quantity) for 2019 for each product and show it using a Pie chart. Understand percentage contribution from each product

```python
plt.style.use('seaborn-notebook')

product = ['FaceCream', 'FaceWash', 'ToothPaste', 'Soap', 'Shampo', 'Moisturizer']
total_Qty = company_data[product].sum()

plt.figure(figsize=(10,10))
plt.pie(total_Qty, labels=total_Qty.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Contribution of Products to Total Sales Quantity')
plt.axis('equal') 
plt.legend(total_Qty.index, loc='best')
plt.show()
```

### 89. Create line plots for Soap & Facewash of all months in a single plot using Subplot
