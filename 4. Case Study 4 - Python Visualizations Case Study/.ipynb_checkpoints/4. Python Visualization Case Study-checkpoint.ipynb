# Case Study on  Visualizations in Python on Sales Data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings ; warnings.filterwarnings('ignore')

import datetime as dt
```

```python
plt.style.use('seaborn-notebook')
```

```python
sales_data = pd.read_csv('SalesData.csv')

sales_data.head(3)
```

# Q1. Compare Sales by region for 2016 with 2015 using bar chart

```python
region_sales = sales_data.groupby(['Region'])['Sales2015','Sales2016'].agg(['sum'])

region_sales.plot(kind='bar', figsize=(18,7), fontsize = 10)
plt.xlabel('Region')
plt.ylabel('Sales')
plt.legend(title='Year')
plt.show()
```

## Q2. What are the contributing factors to the sales for each region in 2016. Visualize it using a Pie Chart.

```python
region_2016 = sales_data.groupby(['Region'])['Sales2016'].agg('sum')

region_2016.plot(kind= 'pie', autopct='%1.1f%%', figsize = (9,9), fontsize = 20)
plt.title('Pie Chart of Sales 2016', fontsize = 20, fontweight= 16)
plt.ylabel('')
plt.show()
```

## Q3. Compare the total sales of 2015 and 2016 with respect to Region and Tiers 

```python
total_sales = sales_data.groupby(['Region','Tier'])['Sales2015', 'Sales2016'].agg('sum')

total_sales.plot(kind= 'bar', figsize= (12,6), fontsize = 20)
plt.xlabel('Tier', fontsize = 20)
plt.ylabel('Sales', fontsize = 20)
plt.title('Comparison of total sales wrt Region and Tiers', fontsize = 25)
plt.legend(title='Year')
plt.show()
```

# Q4. In East region, which state registered a decline in 2016 as compared to 2015?

```python
east_sales_data = sales_data[sales_data['Region'] == 'East']

east_sales_total = east_sales_data.groupby(['State'])['Sales2015','Sales2016'].agg('sum')

east_sales_total.plot(kind= 'bar', figsize=(12,6))
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('Sales comparison for 2015 and 2016 in East region')
plt.show()
```

# Q5. In all the High tier, which Division saw a decline in number of units sold in 2016 compared to 2015? 

```python
hightier_sales_data = sales_data[sales_data['Tier'] == 'High']

hightier_sales_total = hightier_sales_data.groupby(['Division'])['Units2015','Units2016'].agg('sum')

hightier_sales_total.plot(kind= 'bar', figsize=(12,6))
plt.xlabel('Division')
plt.ylabel('Units')
plt.title('Division wise Sales comparison for units sold in 2015 and 2016 in High Tier')
plt.show()
```

# Q6. Create a new column Qtr using numpy.where() or any suitable utility in the imported dataset. The Quarters are based on months and defined as


>  Jan - Mar : Q1
>   Apr - Jun : Q2
>   Jul - Sep : Q3
>   Oct - Dec : Q4

```python
sales_data.Month = pd.to_datetime(sales_data.Month, format='%b')


sales_data['Qtr'] = np.where(sales_data['Month'].dt.month.isin([1,2,3]), 'Q1',
                             np.where(sales_data['Month'].dt.month.isin([4,5,6]), 'Q2',
                                      np.where(sales_data['Month'].dt.month.isin([7,8,9]), 'Q3','Q4')))

sales_data
```

# Q7. Compare Qtr wise sales in 2015 and 2016 in a bar plot

```python
qtr_wise_sale = sales_data.groupby(['Qtr'])['Sales2015','Sales2016'].agg('sum')

qtr_wise_sale.plot(kind='bar')
plt.xlabel('Quater')
plt.ylabel('Sales')
plt.title('Quater Wise Sales in 2015 and 2016')
plt.show()
```

# Q8. Determine the composition of Qtr wise sales in and 2016 with regards to all the Tiers in a pie chart.

```python
df = sales_data.groupby(['Qtr','Tier'])['Sales2016'].agg('sum').unstack()

df[df < 0] = 0

quarters = df.index

for qtr in quarters:
    df.loc[qtr].plot(kind='pie', title=f"Quarter {qtr} Sales Composition", autopct='%1.1f%%')
    plt.ylabel('')
    plt.show()
```
