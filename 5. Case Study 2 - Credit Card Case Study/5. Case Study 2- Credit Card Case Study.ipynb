### 1. In the above dataset

### b. In case spend amount is more than the limit, replace it with 50% of that customer’s limit. (customer’s limit provided in acquisition table is the per transaction limit on his card)

```python
acquisition_df = pd.DataFrame(acquisition)
spend_df = pd.DataFrame(spend)
repayment_df = pd.DataFrame(repayment)

merged_df = pd.merge(spend_df, acquisition_df[['Customer', 'Limit']], on='Customer', how='left')

merged_df['Amount'] = merged_df.apply(lambda row: min(row['Amount'], 0.5 * row['Limit']) if row['Amount'] > row['Limit'] else row['Amount'], axis=1)

print(merged_df)
```

### c. Incase the repayment amount is more than the limit, replace the repayment with the limit.

```python
merged_repayment_df = pd.merge(repayment_df, acquisition_df[['Customer', 'Limit']], on='Customer', how='left')

merged_repayment_df['Amount'] = merged_repayment_df.apply(lambda row: min(row['Amount'], row['Limit']) if pd.notna(row['Amount']) and row['Amount'] > row['Limit'] else row['Amount'], axis=1)

print(merged_repayment_df)
```

### e. If the monthly rate of interest is 2.9%, what is the profit for the bank for each month? (Profit is defined as interest earned on Monthly Profit. Monthly Profit = Monthly repayment– Monthly spend. Interest is earned only on positive profits and not on negative amounts)

```python
spend_df['Month'] = pd.to_datetime(spend_df['Month'], format='%d-%b-%y')
repayment_df['Month'] = pd.to_datetime(repayment_df['Month'], format='%d-%b-%y')

merged_df = pd.merge(spend_df, repayment_df, on=['Customer', 'Month'], how='outer', suffixes=('_spend', '_repayment'))

merged_df['Amount_spend'].fillna(0, inplace=True)
merged_df['Amount_repayment'].fillna(0, inplace=True)

merged_df['Monthly_Profit'] = merged_df['Amount_repayment'] - merged_df['Amount_spend']

interest_rate = 0.029
merged_df['Interest_Earned'] = merged_df['Monthly_Profit'].apply(lambda x: max(0, x * interest_rate))

total_profit_per_month = merged_df.groupby('Month')['Interest_Earned'].sum()

print("Profit for the bank for each month:")
print(total_profit_per_month)
```

### 5. Write user defined PYTHON function to perform the following analysis:

You need to find top 10 customers for each city in terms of their repayment amount by different products and by different time periods i.e. year or month. The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.

```python
def top_customers_analysis(repayment, acquisition, product, time_period):
    """
    Analyze the top 10 customers for each city in terms of repayment amount.
    
    Parameters:
    - repayment (DataFrame): DataFrame containing repayment data.
    - acquisition (DataFrame): DataFrame containing acquisition data.
    - product (str): Product type (Gold/Silver/Platinum).
    - time_period (str): Time period for analysis ('yearly' or 'monthly').
    
    Returns:
    - DataFrame with the top 10 customers for each city, product, and time period.
    """
    merged_data = pd.merge(repayment, acquisition, on='Customer', how='inner')

    filtered_data = merged_data[merged_data['Product'] == product]

    if 'City' not in filtered_data.columns:
        raise ValueError("The 'City' column is not present in the merged data.")

    
    if time_period == 'yearly':
        filtered_data['Year'] = pd.to_datetime(filtered_data['Month']).dt.year
    elif time_period == 'monthly':
        filtered_data['Month'] = pd.to_datetime(filtered_data['Month'])
        filtered_data['Year'] = filtered_data['Month'].dt.year
        filtered_data['Month'] = filtered_data['Month'].dt.month

    
    grouped_data = filtered_data.groupby(['City', 'Year', 'Customer'])['Amount'].sum().reset_index()

    
    sorted_data = grouped_data.sort_values(by='Amount', ascending=False)

    
    top_customers = sorted_data.groupby(['City', 'Year']).head(10)

    return top_customers




product_type = 'Gold'
time_period = 'yearly'
top_customers_result = top_customers_analysis(repayment, acquisition, product_type, time_period)

print(top_customers_result)
```
