```python
import numpy as np
import pandas as pd

import matplotlib as plt
import seaborn as sns

import datetime as dt
from datetime import datetime

import warnings; warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt 

import statistics
import pandas_profiling 
from pandas_profiling import ProfileReport

from sklearn.model_selection import train_test_split

from statsmodels.api import OLS
```

## Import claims_data.csv and cust_data.csv which is provided to you and combine the two datasets appropriately to create a 360-degree view of the data. Use the same for the subsequent questions.

```python
claims = pd.read_csv('claims.csv')
cust_data = pd.read_csv('cust_demographics.csv')

insurance_claims_data = pd.merge(left=claims,
                                right=cust_data,
                                how='inner',
                                left_on='customer_id',
                                right_on='CUST_ID')

insurance_claims_data
```

## Perform a data audit for the datatypes and find out if there are any mismatch within the current datatypes of the columns and their business significance.

```python
data_types = insurance_claims_data.dtypes
print("Data Types:")
print(data_types)

insurance_claims_data['claim_date'] = pd.to_datetime(insurance_claims_data['claim_date'], errors='coerce')
insurance_claims_data['DateOfBirth'] = pd.to_datetime(insurance_claims_data['DateOfBirth'], errors='coerce')

print("\nUpdated Data Types:")
print(insurance_claims_data.dtypes)
```

## Convert the column claim_amount to numeric. Use the appropriate modules/attributes to remove the $ sign.

```python
insurance_claims_data['claim_amount'] = pd.to_numeric(insurance_claims_data['claim_amount'].str.replace('[\$,]', '', regex=True), 
                                                      errors='coerce')

print(insurance_claims_data['claim_amount'])
```

## Of all the injury claims, some of them have gone unreported with the police. Create an alert flag (1,0) for all such claims.

```python
insurance_claims_data['unreported_injury_claim'] = (
    (insurance_claims_data['claim_type'] == 'Injury only') &
    (insurance_claims_data['police_report'].isin(['No', 'Unknown', 'Yes']))
).astype(int)


print(insurance_claims_data[['claim_id', 'claim_type', 'police_report', 'unreported_injury_claim']])
```

## One customer can claim for insurance more than once and in each claim, multiple categories of claims can be involved. However, customer ID should remain unique. Retain the most recent observation and delete any duplicated records in the data based on the customer ID column.

```python
insurance_claims_data.sort_values(by='claim_date', ascending=False, inplace=True)

insurance_claims_data.drop_duplicates(subset='customer_id', keep='first', inplace=True)
insurance_claims_data.drop('CUST_ID', axis=1, inplace=True)
insurance_claims_data.reset_index(drop=True, inplace=True)

insurance_claims_data
```

## Check for missing values and impute the missing values with an appropriate value. (mean for continuous and mode for categorical)

```python
missing_values = insurance_claims_data.isnull().sum()
print("Missing Values:")
print(missing_values)




for column in insurance_claims_data.columns:
    if insurance_claims_data[column].dtype == 'float64':
        insurance_claims_data[column].fillna(insurance_claims_data[column].mean(), inplace=True)
        
    elif insurance_claims_data[column].dtype == 'object':
        insurance_claims_data[column].fillna(insurance_claims_data[column].mode()[0], inplace=True)

print("\nDataFrame after imputing missing values:")
insurance_claims_data
```

## Calculate the age of customers in years. Based on the age, categorize the customers according to the below criteria
Children < 18
Youth 18-30
Adult 30-60
Senior > 60

```python
current_date = datetime.now()

future_birth_years = insurance_claims_data['DateOfBirth'].dt.year > current_date.year
insurance_claims_data.loc[future_birth_years, 'DateOfBirth'] -= pd.DateOffset(years=100)

insurance_claims_data['age'] = (current_date - insurance_claims_data['DateOfBirth']).dt.days // 365

current_date_formatted = current_date.strftime('%Y-%m-%d')

bins = [0, 18, 30, 60, float('inf')]
labels = ['Children', 'Youth', 'Adult', 'Senior']

insurance_claims_data['age_category'] = pd.cut(insurance_claims_data['age'], bins=bins, labels=labels, right=False)



print(insurance_claims_data[['customer_id', 'DateOfBirth', 'age', 'age_category']])
print("Current Date:", current_date_formatted)
```

## What is the average amount claimed by the customers from various segments?

```python
average_claim_amount_by_segment = insurance_claims_data.groupby('Segment')['claim_amount'].mean()

print("Average Claim Amount by Segment:")
print(average_claim_amount_by_segment)
```

```python
import scipy.stats as stats
from scipy.stats import chi2_contingency

import datetime
```

## Is there any similarity in the amount claimed by males and females? 

```python
males_claim_amount = insurance_claims_data.loc[insurance_claims_data['gender'] == 'Male', 'claim_amount']
females_claim_amount = insurance_claims_data.loc[insurance_claims_data['gender'] == 'Female', 'claim_amount']

# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(males_claim_amount, females_claim_amount, equal_var=False)

# Check the p-value
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant difference.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference.")
```

## Is there any relationship between age category and segment?

```python
contingency_table = pd.crosstab(insurance_claims_data['age_category'], insurance_claims_data['Segment'])

# Perform Chi-square test of independence
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Check the p-value
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant relationship.")
else:
    print("Fail to reject the null hypothesis. There is no significant relationship.")
```

## The current year has shown a significant rise in claim amounts as compared to 2016-17 fiscal average which was $10,000.

```python
current_year = datetime.datetime.now().year

current_year_claim_amount = insurance_claims_data.loc[insurance_claims_data['claim_date'].dt.year == current_year, 'claim_amount']
average_claim_2016_17 = 1000 

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(current_year_claim_amount, average_claim_2016_17)

# Check the p-value
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant rise.")
else:
    print("Fail to reject the null hypothesis. There is no significant rise.")
```

## Is there any difference between age groups and insurance claims?

```python
age_groups = ['Children', 'Youth', 'Adult', 'Senior']
data_by_age_group = [insurance_claims_data.loc[insurance_claims_data['age_category'] == group, 'claim_amount'] for group in age_groups]

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(*data_by_age_group)

# Check the p-value
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant difference.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference.")
```

## Is there any relationship between total number of policy claims and the claimed amount?

```python
total_policy_claims = insurance_claims_data['total_policy_claims']
claimed_amount = insurance_claims_data['claim_amount']

# Perform Pearson correlation analysis
correlation_coefficient, p_value = stats.pearsonr(total_policy_claims, claimed_amount)

# Check the p-value
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is a significant linear relationship.")
else:
    print("Fail to reject the null hypothesis. There is no significant linear relationship.")
```
